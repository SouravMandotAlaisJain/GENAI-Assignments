```python
import nltk
import random
from collections import defaultdict

nltk.download('punkt_tab')

file_name = "t8.shakespeare.txt"
with open(file_name, 'r', encoding='utf-8') as file:
    text_data = file.read()

words = nltk.word_tokenize(text_data)

markov_chain = defaultdict(list)

for current_word, next_word in zip(words[:-1], words[1:]):
    markov_chain[current_word].append(next_word)

def generate_text(chain, start_word, max_words=100):
    """Generate with Markov Chains"""
    current_word = start_word
    text = [current_word]

    for _ in range(max_words - 1):
        if current_word not in chain or not chain[current_word]:
            break
        next_word = random.choice(chain[current_word])
        text.append(next_word)
        current_word = next_word

    return ' '.join(text)

start_word = random.choice(words)

print("Generated Text:")
print(generate_text(markov_chain, start_word))
```
